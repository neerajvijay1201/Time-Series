# using moving average method

```python
moving_averages = []

for i in range(len(y) -9):
    window = y[i:i+10]
    average = sum(window) / len(window)
    moving_averages.append(average)

print("Original List:", y)
print("Moving Averages (Window Size 4):", moving_averages)
```

```python
len(moving_averages)
```

```python
final_moving_averages=moving_averages+[0,0,0,0,0,0,0,0,0]
print(final_moving_averages,end=',')
```

```python
plt.plot(x,y)
plt.plot(x,final_moving_averages)
```
